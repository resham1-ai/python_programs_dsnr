````python
# write a program to find the sum of all even numbers in between 1 to 100 using while loop 
# Initialize variables
num = 1
sum_even = 0

# Loop through numbers from 1 to 100
while num <= 100:
    if num % 2 == 0:  # Check if the number is even
        sum_even += num
    num += 1  # Increment the number

# Display the result
print("The sum of all even numbers between 1 and 100 is:", sum_even)

````

````python
# write a program to find the product of all numbers in between 10 to 30 using while loop 
# Initialize variables
num = 10
product = 1

# Use while loop to multiply numbers from 10 to 30
while num <= 30:
    product *= num
    num += 1

# Display the result
print("The product of all numbers from 10 to 30 is:", product)
````

````python
# write a program to find the sum of all odd numbers and product of all even numbers in between 1 to 100 using while loop
# Initialize variables
num = 1
sum_odd = 0
product_even = 1

# Loop from 1 to 100
while num <= 100:
    if num % 2 == 0:
        product_even *= num
    else:
        sum_odd += num
    num += 1

# Display the results
print("Sum of all odd numbers between 1 and 100:", sum_odd)
print("Product of all even numbers between 1 and 100:", product_even)
````
